```python
#############
import pandas
import requests
import json
from bs4 import BeautifulSoup
t_list=[]
for pg in range(1,116):
    URL     ="http://www.books.com.tw/web/books_bmidm_0102/?o=1&v=1&page=%s"%pg
              
    resp    = requests.get(URL)
    soup    = BeautifulSoup(resp.text,'html5lib')
    selects = soup.select('div.msg > h4 > a')
    for s in selects:
        g = s.get('href')
        print(g)
        t_list.append(g)
with open('books_literatural_Chinese_writing.csv','w') as f:
    f.write(json.dumps(t_list))
```
