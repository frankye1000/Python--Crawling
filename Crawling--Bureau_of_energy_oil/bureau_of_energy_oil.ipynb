```python
##date
import requests
import re
import pandas as pd 
import sqlalchemy
import concurrent.futures
from concurrent.futures import ThreadPoolExecutor
from bs4 import BeautifulSoup

page=[]
for i in range(127,933):
    page.append(i)

def enrge(pg):
    payload={
        'includes':'1',
        'oilid':'01',
        'oilname':'(unable to decode value)',
        'category':'wDetail',
        'SearchCaption':'(unable to decode value)',
        'GroupField':'are_sample',
        'FieldName':'(unable to decode value)',
        'SearchStatus':'Real',
        'PriceStatus':'Re',
        'weekid':'%s'%pg
        }

    energe_dict={}
    resp = requests.post('https://www2.moeaboe.gov.tw/oil102/oil1022010/A01/A0101/oilsearch_cpkw.asp',data=payload)
    soup = BeautifulSoup(resp.text,'html5lib')
    selects = soup.select("table > tbody > tr > td ")
    
    dates  = selects[0].text.encode('ISO-8859-1').decode('big5')
    date   = re.findall("週別時間：(.+)",dates)[0]
    energe_dict["date"]  = date
    return energe_dict
    
date_list=[]
with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    for data in executor.map(enrge, page):
        date_list.append(data)   
df_date = pd.DataFrame(date_list)
```
